Is the group significantly different (with respect to systolic blood pressure!) from the regular population?

### Hypothesis Testing
Helps us realize if the sample data is really representative for the rest of the population

#### Null Hypothesis (H0):
For our specific case it would mean to assume that the mean systolic blood pressure of the sample equals the one of the whole population (120 mm). 

#### Alternate hypothesis (H1 / Ha):
The oposite of H0.

#### Significance Level: 
Indicates if we are confident enough to reject the null hypothesis

#### Test statistic:
##### Step 1&2: setting the hypothesis; We compare H1 with H0
H0: μ = 120 mm Hg
H1: μ ≠ 120 mm Hg

There is a significant difference between sample mean and population mean systolic blood pressure.

##### Step 3: One-tailed or a two-tailed test?
Two-tailed test.

The sample mean could be differ in both negative or positive directions. If it was a one tailed test, our H1 would say μ < 120 mm Hg


##### Step 4: Deciding the statistic test based on the information available
Our sample size is 100 which is why we will use Z-test and not T-test to test the null hypothesis.

##### Step 5: Level of significance and critical value (Z-Value)



Significance level 0.05. This means we are willing to accept a 5% chance of making an error.

Finding the Z critical value:
1) Dividing the significance level (α) by 2


2) Substract α/2 from 1


3) Search the value in the Z table. 


Z critical value = 1.9 + 0.06 = 1.96 (+- since it's a two-tailed test)

We fail to reject the null hypothesis, as the critical value is not extreme enough. 

We will continue to investigate if the null hypothesis is true or not. If the calculated test will fall outside of the critical value's range, then we have evidence to reject the null hypothesis

- The standard deviation of our sample should be equal to the standard deviation of the population if H0 was true. In the formula below we approximate the population standard deviation equals the sample standard deviation


- Z-score tells us how far away we are from the average. What is the probability of getting 130.1 mm Hg with a standard deviation of 2.121?


```python
### import math

sample_mean = 130.1
pop_mean = 120
pop_std = 21.21
n = 100

statistic = (sample_mean - pop_mean)/(pop_std/math.sqrt(n))
statistic
```
